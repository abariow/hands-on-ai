```python
import os
```

```python
# Read all the emails and separate them into spam and ham
spams, hams = [], []
for folder_name, subfolders, file_names in os.walk('data'):
    for file_name in file_names:
        file_path = os.path.join(folder_name, file_name)
        with open(file_path, 'rb') as f:
            if 'ham' in file_path:
                hams.append(f.read().decode('utf-8', errors='ignore'))
            elif 'spam' in file_path:
                spams.append(f.read().decode('utf-8', errors='ignore'))
                
# Remove duplicates
spams = list(set(spams))
hams = list(set(hams))

print('Number of spams: ', len(spams))
print('Number of hams: ', len(hams))
```
